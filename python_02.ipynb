### 1. List comprehension

#### 1.1. Что такое list_comprehension

Рассмотрим создание массива в Python:

````python
a = []
for i in range(10):
    a.append(i)
a
````

Но в python можно создать массив так:

````python
b = [i for i in range(10)]
b
````

Написанное можно воспринимать примерно так:

"прогоним такой `for`, на каждой итерации возьмём это самое `i`, из них породим массив (список, конечно, на самом деле)"


Не обязательно брать именно i на каждой итерации - с его участием можно выполнить любую операцию.

Здесь, например, список заполнится квадратами целых чисел:

````python
c = [i**2 for i in range(10)]
print(c)
````

Мы можем использовать не каждое `i`:

````python
# Только квадраты кратных 3 чисел
# Пример для демонстрации средств языка. Да, можно сделать эффективнее.
d = [i**2 for i in range(10) if i % 3 == 0]
print(d)
````

Собственно говоря, `i` может и вовсе не участвовать в том выражении, которое пойдёт в список.

Например, зальём список единицами

````python
f = [1 for i in range(10)]
print(f)
````

Если `i` не участвует в расчёте, его можно опустить (кстати, работает и для обычного `for`)

````python
g = [2 for _ in range(10)]
print(g)
````

#### 1.2. Проблема с mutable и копированием

Конкретно заливку в прошлом примере можно было проще сделать:

````python
h = [3] * 10
print(h)
````

Теперь вспомним о mutable типах и подумаем, почему к примеру двумерный массив или массив словарей так лучше не заливать.

````python
k_mutable = [[0, 0]] * 2
print("Array before:", k_mutable)
print("k[1][0] before:", k_mutable[1][0])

# Попытаемся изменить только элемент [0, 0]
k_mutable[0][0] = 1
print("Array after:", k_mutable)
print("k[1][0] after:", k_mutable[1][0])
````

Для таких случаев list_comprehension - один из валидных способов создания:

````python
k = [[0, 0] for _ in range(2)]
print("Array before:", k)
print("k[1][0] before:", k[1][0])

# Попытаемся изменить только элемент [0, 0]
k[0][0] = 1
print("Array after:", k)
print("k[1][0] after:", k[1][0])
````

#### 1.3. Ввод массивов.

Довольно часто удобно использовать list comprehension для разбора входа.

Например, вот так можем получить введённый с консоли массив целых чисел в виде именно целых чисел (не строк):

````python
a = [int(s) for s in input().split()] # input: 3 4 5
print(a)
````

#### 1.4. Set и dict comprehension.

Аналогичный comprehension можно использовать и для `set`-ов и `dict`-ов

__ВНИМАНИЕ!__ Контейнеры `set` и `dict` в Python не упорядоченные! Здесь можно получить некоторые сюрпризы.


````python
a = set(int(s) for s in [1, 1, 2, 3, 2])
print(a)
````

Этот пример довольно искусственный. Но это `dict`-comprehension.

````python
d = {i: i**3 for i in range(5)}
print(d)
````

### 2. Iterable.

#### 2.1. Примеры iterable

В Python-е есть довольно фундаментальное понятие - iterable.

Всё, что может последовательно перебираться - iterable.

Очень многие функции на вход на самом деле принимают именно iterable.

Например, есть у нас `list` некоторой длины, он iterable:

````python
a = [1, 2, 3, 4]
````

Поэтому к нему, например, можно применить max - max принимает на вход iterable сущность произвольной длины

````python
print(max(a))
````

Аналогично set - он тоже iterable

````python
s = {1, 2, 3}
print(max(s))
````

Более того, выполнение некоторого кода тоже порождает iterable результат

И этот iterable результат можно использовать "на лету", передавая куда-нибудь без создания промежуточных коллекций

````python
print(max(i**2 for i in range(10)))
````

#### 2.2. Операции над iterable.

##### Enumerate

Посмотрим, что ещё можно сделать с iterable (здесь характерные примеры).

Создадим set из символов строки (просто потому что можем - ну и для ещё одного примера кастов, конечно).

````python
s = set("abcd")
````

Просто `for`-ы уже были (примерно аналоги range for-ов в C++).

А если iterable хочется обойти, имея не только значение, но и индекс, то можно использовать enumerate (Ещё раз напомним, что `set` и `dict` в Python не упорядоченные!)

````python
for index, value in enumerate(s):
    print(f'Value {index}: {value}')
````

Для подобного перечисления `dict` нужен метод `items()`.

##### Распаковка iterable:

Получить значения из iterable можно так:

````python
il = [1, 2, 3, 4]

# Под именем il у нас сам list
print("List itself:", il, type(il))
# А обратившись *il можно получить все значения из него
print("Values from list:", *il)
````

То же можно сделать с другими iterable, но при распаковки переменной больше одного раза возможны сюрпризы:

К примеру, `range`:

````python
rl = range(4)

print("Range itself:", rl, type(rl))
print("Values from range:", *rl)
print("Values from range again:", *rl)
````

И для результатов list_comprehension:

````python
lc = (i ** 2 for i in range(4))

print("Object itself:", lc, type(lc))
print("Values:", *lc)
print("Values again:", *lc)
````

##### Параллельная распаковка iterable:

Если мы захотим параллельную итерацию по двум спискам, её можно организовать вот так:

````python
for i, s in zip(il, rl):
    print(f'Ad-hoc pair: {i} {s}')
````

Технически `zip` возвращает (сюрприз!) итератор. (Да, ровно как в C++. Иногда они возвращаются.)

````python
print(type(zip(il, rl)))
````

При желании этот итератор можно к чему-нибудь кастануть.

````python
print(type(list(zip(il, rl))))
````

### 3. Более подробно о списках.

#### 3.1. Операции над списками: добавление и удаление элементов.

Создадим список:

````python
a = [i for i in range(10)]
a
````

В него можно добавить элементы:

````python
a.append(10)
a.insert(0, 42)
a
````

````python
a.append(b)
a
````

#### 3.2. Индексация в Python: slices. 

И снова у нас есть список (на самом деле опять iterable сущность, но пусть будет список для примера)

````python
a = [i for i in range(10)]
a
````

Мы можем выделить его диапазон (slice) и обратиться к нему:

````python
a[1:2] = []
a
````

##### Индексы у slice.

Ещё можно делать индексацию по отрицательным индексам - с конца.

````python
# Это будет первый элемент
print(a[0])
# А это последний
print(a[-1])
````

Слайсы можно задавать аналогично (если пропустить какую-то из цифр - там будет "от начала" или "до конца" соответственно).

````python
print(a[-3:])
````

Еще можно получить слайс с заданной позиции с заданным шагом:

````python
b = [i for i in range(10)]
print(b)
print(b[1::2])
````

И в обратную сторону так ходить тоже можно:

````python
print(b[::-1])
````

### 4. Tuples.

#### 4.1. Что такое tuple?

На первый взгляд tuple почти не отличается от списка, только скобочки круглые.

На них работает всё то, что было выше для списков:


````python
t = (1, 2, 3)

print(t)
print(t[0])
````

Но tuple - неизменяемая, и присвоить значение по индексу не возможно - только заменить значение полностью

````python
t[0] = -1
````

#### 4.2. Распаковка и запаковка.

При создании можно опускать скобки.

````python
tt = 1, 2, 3
print(tt)
````

Дальше начинается лёгкая магия.

Можно распаковать содержимое тапла в несколько переменных - это часто удобно, но это ладно.

````python
a, b, c = tt
print(a, b)
````

А теперь объединим эти два примера:

````python
b, a = a, b
print(a, b)
````

Последний пример работает на самом деле так:

- то, что справа от равно запаковывается во временный тапл,
- этот тапл присваивается таплу слева от равно,
- тапл слева от равно распаковыывается в именованные переменные.

А синтаксис выглядит лёгкой магией, потому что скобочки у таплов можно не писать.

#### 4.3. Tuple как возвращаемое значение.

Ещё на tuples работает такая особенность функций в Python-е:

У нас есть функция, она принимает пару параметров (это ок), и потом внезапно возвращает два значения через запятую (это как?)

````python
def func(a, b):
    return a + b, a - b
````

На самом деле функция возвращает тапл, который можно принять и распаковать:


````python
x, y = func(1, 2)
print(x)
print(y)
````

Ну или не распаковать, а так и оставить таплом:


````python
q = func(1, 2)
print(q)
````

Но если уж начали распаковывать, то делать это надо с умом:


````python
i, j, k = func(1, 2)
````

### 5. Работа с файлами.

Базовая работа с файлами может быть устроена примерно так:

````python
# Откроем test.txt на чтение ('r' от 'read')
f = open("test.txt", 'r')
# Прочитаем весь файл разом в одну строку, тут же выведем её на экран
print(f.read())
# Закроем файл
f.close()
````

Но чуть более красиво написать так:

````python
with open("test.txt", 'r') as f:
    print(f.read())
````

Это такой типовой блок:

```
with ... as ...:
    ...
```

Мы пытаемся захватить ресурс (открыть файл в данном случае) и выполняем тело блока только в том случае, если захватить ресурс удалось.

В конце блока ресурс освободится (в данном случае файл закроется).


### 6. Исключения.

Исключения в Python-е есть, как же без них.

Рассмотрим такой код:

````python
a = [int(s) for s in ["1", "2", "three"]]
````

Хотелось бы в таких случаях не умирать на месте, а иметь шанс обработать ошибку - для этого нужен try и ловля исключений, как и в других языках.


К примеру такой код не упадёт, а попадёт в блок except, в котором можно написать свою логику.

````python
try:
    a = [int(s) for s in ["1", "2", "three"]]
except:
    pass # pass - служебная конструкция со смыслом "этот блок пустой, и это так задумано"
````

Мы можем передать исключение внутрь блока `except`:

````python
try:
    a = [int(s) for s in ["1", "2", "three"]]
except Exception as e:
    print(e)
````

При этом можно отлавливать отдельные типы ошибок и кидать пойманные исключения:

````python
try:
    a = [int(s) for s in ["1", "2", "three"]]
except RecursionError:
    print("Where is recursion???")
except ValueError as e:
    print("I got ValueError")
    raise e
````

### 7. Ключи для сортировки.

В языке Python есть встроенная сортировка - посмотрим, как мы можем поменять ключ, по которому значения сортируются.

Пусть у нас есть вот такие данные - список таплов.
К примеру, точки, поля каждой это (x, y, color).


````python
a = [(1,2,'red'), (1,0,'blue'), (3,5,'red')]
a
````

Для построения ключа, по которому значения сортируются, есть аргумент `key` функции `sorted`:

````python
print(sorted(a, key = lambda e: e[0])) # Сорировка по x
````

Значением ключа может быть `tuple` - сравниваются сначала первые значения, потом вторые и т.д... Так, если мы хотим отсортировать сначала по X, а потом по Y:

````python
print(sorted(a, key = lambda e: (e[0], e[1])))
````

Ещё тут стоит сказать, что key не требует строго лямбду, и подойдёт просто функция, которая для элемента вернёт ключ.

Просто часто проще всего в роли этой функции сразу на месте лямбду и написать.

````python
# Функция, на вход принимает очередной элемент (тапл),
# распаковывает его (потому что может), в роли ключа возвращает цвет
def color_of_point(p):
    _, _, color = p
    return color

# Примерно аналогично, только ключ не равен какому-то полю, а вычисляется
def vector_square_length(p):
    x, y, _ = p
    return x**2 + y**2

# А теперь в роли ключа вернём тапл, собрав его вызовами прошлых функций
def two_param(p):
    return color_of_point(p), -vector_square_length(p)
````

Сортировка по цвету

````python
print(sorted(a, key = color_of_point))
````

Сортировка по удалённости от координат

````python
print(sorted(a, key = vector_square_length))
````

Творческая сортировка по паре параметров

````python
print(sorted(a, key = two_param))
````

### 8. Глобальные переменные.

Работа с глобальными переменными в Python несколько специфична.

````python
# Пусть у нас есть глобальная переменная.
# Это всегда так себе идея. Но пусть есть.
TEST = 1
````

Эта функция хочет читать глобальную переменную - имеет право.


````python
def func1():
    print("Inside func1", TEST)

print("Initial value", TEST)
func1()
````

Эта функция хочет писать в глобальную переменную.

И как будто ей дали это сделать. Но нет.

````python
def func2():
    TEST = 10
    print("Inside func2", TEST)

print("Before func2 call", TEST)
func2()
print("After func2 call", TEST)
````

А вот эта функция правда сможет писать в глобальную переменную.

````python
def func3():
    global TEST
    TEST = 10
    print("Inside func3", TEST)

print("Before func3 call", TEST)
func3()
print("After func3 call", TEST)
````
