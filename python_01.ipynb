### 4. Конструкции языка.

#### 4.1. If/else

Без `if`-ов жить, конечно, нельзя. В Python-е они выглядят примерно так.

```python
a = 3

if a > 0:
    print("positive")
else:
    if a < 0:
        print("negative")
    else:
        print("zero")
```

И, чтобы упростить `if...else...if...else...` конструкции, как приведённая выше, есть `elif`:

```python
a = 0

if a > 0:
    print("positive")
elif a < 0:
    print("negative")
else:
    print("zero")
```

#### 4.2. Циклы.

В python-е в отличие от C `for` итерируется не через счётчик, а по элементам объекта (коллекции) (C++ так тоже может). Но для циклов по диапазону чисел (от 0 до n, от a до b, с шагом) есть объекты `range` позволяющие описывать цикл просто:

От 0 до n:

```python
n = 6
for i in range(n):
    print(i)
```

От a до b:

```python
a = 2
b = 6

for i in range(a, b):
    print(i)
```

И если добавить шаг c:

```python
a = 2
b = 6
c = 2

for i in range(a, b, c):
    print(i)
```

Можно использовать и объекты кроме `range`, но об этом позже.

Цикл `while` же работает обычным образом:

```python
a = 0
while a < 10:
    print(a)
    a += 4
```

#### 4.3. Функции.

Если не вдаваться в особенности (именованные параметры, дефолтные значения параметров), функции в Python описываются так:

```python
def fact(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res

n = 5
print("%d! = %d" % (n, fact(n)))
```

#### 5.2. Словари.

Для хранения составных данных есть приличное количество вариантов. Возьмём для старта и для примера `dict`. Это примерно как map в C++, хотя, конечно, не совсем.

Пусть у нас есть три человека. У каждого есть имя и возраст.

```python
user_a = {
    'name': "A",
    'age': 20
}
user_b = {
    'name': "UserB",
    'age': 17
}
user_c = {
    'name': "C",
    'age': 25
}
```

Сделаем из них массив

```python
users = [user_a, user_b, user_c]
```

Обойдём людей, отсортировав их по возрасту:

```python
for user in sorted(users, key = lambda u: u['age']):
    print("User %s, age %d" %(user['name'], user['age']))
```

Отсортируем людей по имени, меняя исходный массив. Потом обойдём отсортированное:

```python
users.sort(key = lambda u: u['name'])
for user in users:
    print("User %s, age %d" % (user['name'], user['age']))
```

#### 5.6. И о mutable/immutable типах и ссылках.

В C++ было много нюансов с передачей параметров в функции. Можно было передавать по значению, по указателю, по ссылке. И ещё были const-ы.

Как это устроено в Python-е? Несколько иначе. Бывают неизменяемые типы данных (immutable) и изменяемые (mutable). Всё immutable передаётся по значению. Всё mutable - по ссылке.

К примеру, `list`, `dict` и `set` - изменяемые типы:

```python
def change_list(l):
    l.append(42)
    l[0] = -1

my_list = [1, 2, 3]
print(my_list)
change_list(my_list)
print(my_list)
```

А `int`, `float`, `string` - нет:

```python
def change_string(s):
    s += " test"
    print("Inside function: %s" % s)

my_string = "string"
print(my_string)
change_string(my_string)
print(my_string)
```

По этой же причине мы не можем менять элементы строки как у массива:

```python
print("list before:", my_list)
my_list[0] = -42
print("list after:", my_list)

# Сделаем то же самое со строкой:
my_string[0] = "w"
```

Строго говоря, всё передаётся "по присваиванию" (by assignment). То есть вообще-то всё всегда передаётся по ссылке, а сама ссылка - по значению.

Далее immutable сущности не могут быть изменены вообще никак. А mutable могут быть изменены (потому что ссылка), но не могут быть переприсвоены (потому что ссылка по значению).

Но вспоминать об этих тонкостях стоит только тогда, когда хотите странного (или уже его сотворили и отлаживаете).

Хотя со случайным присваиванием ссылки без копирования содержимого часто можно выстрелить себе в ногу:

```python
print("list before:", my_list)

# This is new list, unrelated to the old one, right?
new_list = my_list

new_list[0] += 100
print("list after:", my_list)
```

__Важно!!!__ Присваивание mutable типов присваивает ссылку, не значение, на которое она указывает - изменение элементов `list`, `dict` или вызов меняющих значение методов изменит исходную переменную. Если мы хотим копироват содежимое, это нужно указывать отдельно.

```python
print("list before:", my_list)

new_list = my_list.copy()

new_list[0] += 10
print("list after:", my_list)
```
